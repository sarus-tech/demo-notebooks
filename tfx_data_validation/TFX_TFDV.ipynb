## Let's generate statistics

```
import tensorflow_data_validation as tfdv

print("Script to test TFX Schema inference")
print("TFDV version: {}".format(tfdv.version.__version__))

csv_path = "./adult.csv"
train_stats = tfdv.generate_statistics_from_csv(data_location=csv_path)
i = tfdv.visualize_statistics(train_stats)
print(i)
```

## And now we can compute a schema

```
schema = tfdv.infer_schema(statistics=train_stats)
tfdv.display_schema(schema=schema)
```

```
from google.protobuf.json_format import MessageToJson

jsonSchema = MessageToJson(schema)
print(jsonSchema)
```
